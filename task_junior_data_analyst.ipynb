```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import statsmodels.api as sm
import numpy as np

pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', None)


df = pd.read_excel('data.xlsx')



df['has_deposit'] = df['has_deposit'].map({'yes': 1, 'no': 0}) 


#cleaning data


# print(df)

# print(df.columns)   #checking columns
# print(df.info())   #checking general info
# print(df.describe())  # getting to know data
# print(df.dtypes) #checking data types


# Checking for missing values
# print(df.isnull().sum()) # no empty values

# Dropping duplicate rows, if any
df = df.drop_duplicates()

# print((df == 'unknown').sum()) # unknown=0

```

```python
# Visualizing variables

sns.countplot(data=df, x='age', hue='has_deposit')
plt.show() 

sns.boxplot(x='has_deposit', y='age', data=df)
plt.show()

sns.countplot(data=df, x='emp_var_rate', hue='has_deposit')
plt.show()

sns.countplot(data=df, x='euribor3m', hue='has_deposit')
plt.show()

sns.countplot(data=df, x='pdays', hue='has_deposit')
plt.show()

sns.countplot(data=df, x='has_deposit', hue='marital')
plt.show()


#---------------------------------------------------------------------------

#creating pivot tables
dpivot = pd.read_excel('data.xlsx') 
dpivot['has_deposit'] = dpivot['has_deposit'].map({'yes': 1, 'no': 0})

# pivot -> loan, housing and has_deposit as value
pivot_dpivot = dpivot.pivot_table(index='loan', columns='housing', values='has_deposit', aggfunc='sum', fill_value=0)
print(pivot_dpivot)

# pivot -> marital, job and has_deposit as value
pivot_dpivot = dpivot.pivot_table(index='marital', columns='job', values='has_deposit', aggfunc='sum', fill_value=0)
print(pivot_dpivot)

# pivot -> marital, job and has_deposit as value percent of all marital and job values
pivot_dpivot = dpivot.pivot_table(index='marital', columns='job', values='has_deposit', aggfunc=lambda x: (x.sum() / len(x)) * 100, fill_value=0)
print(pivot_dpivot)

# pecentage view gives us more accurate information


```
